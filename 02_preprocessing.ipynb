# Mengubah data mentah menjadi format yang bisa "dibaca" model (menambahkan Lag, MA, Diff).

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('data/raw/alpenbank_macro_raw.csv', parse_dates=['Date'], index_col='Date')
```

# Lags (Indikator Masa Lalu)

```python
cols_to_lag = ['CPI_YoY', 'SNB_Policy_Rate', 'GDP_Growth']
for col in cols_to_lag:
    for lag in [1, 3, 6]:
        df[f'{col}_Lag_{lag}'] = df[col].shift(lag)
```

# Movingg Avarage

```python
df['SNB_Rate_MA3'] = df['SNB_Policy_Rate'].rolling(window=3).mean()
df['CPI_MA6'] = df['CPI_YoY'].rolling(window=6).mean()
```

# Seasonality Extraction (Untuk Loan Disbursement)

```python
decomp = seasonal_decompose(df['Loan_Disbursement'], model='additive', period=12)
df['Loan_Seasonality'] = decomp.seasonal
```

# Differencing (Perubahan Bulanan)

```python
df['CPI_Change'] = df['CPI_YoY'].diff()
```

```python
df_clean = df.dropna() 
df_clean.to_csv('data/processed/alpenbank_ready_for_model.csv')
print("Selesai: Feature Engineering di 'data/processed/'")
```
